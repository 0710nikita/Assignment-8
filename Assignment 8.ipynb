```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
iris = datasets.load_iris()
```

```python
X_reduce = PCA(n_components = 3).fit_transform(iris.data)
Y = iris.target

fig = plt.figure(1, figsize = (10,8))
axes = Axes3D(fig, elev = -160, azim = 120)
axes.scatter(X_reduce[:, 0], X_reduce[:, 1], X_reduce[:, 2], c = Y,cmap = plt.cm.rainbow_r)

axes.set_title("PCA Directions")
axes.set_xlabel("Eigenvector 1")
axes.set_ylabel("Eigenvector 2")
axes.set_zlabel("Eigenvector 3")

plt.show()
```
